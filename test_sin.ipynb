```python
def error_rate(data,recon_data):
    rd = recon_data.copy()
    rd[rd > 1.0] = 1.0
    rd[rd < -1.0] = -1.0
    return "L1: {0:12.5f} -- L2: {1:12.5f}".format(np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)) , np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)

def view_tree(tree,vecs,vals,descs=None,tree_desc=""):
    viewer_files.write_tree_viewer("temp.pickle",tree,vecs,vals,descs,tree_desc)
    %run tree_viewer.py temp.pickle
```

```python
if CASE == 1:
    ROWS = 400
    COLUMNS = 200
    means_matrix = np.zeros([ROWS,COLUMNS])
    for i in xrange(ROWS):
        for j in xrange(COLUMNS):
            ii = i*2.0*np.pi/ROWS
            jj = j*2.0*np.pi/COLUMNS
            means_matrix[i,j] = np.sin((ii+jj+2*ii*jj)/2.0)*1.0
```

```python
if CASE == 2:
    ROWS = 200
    COLUMNS = 100
    row_data = np.sin(np.arange(ROWS)*2.0*np.pi/ROWS)
    col_data = np.sin(np.arange(COLUMNS)*2.0*np.pi/COLUMNS)
    means_matrix = np.outer(row_data,col_data)
    cplot(means_matrix)
```

```python
if CASE == 3:
    ROWS = 200
    COLUMNS = 200
    means_matrix = np.zeros([ROWS,COLUMNS])
    for i in xrange(ROWS):
        for j in xrange(COLUMNS):
            ii = (i*1.0/ROWS - 0.5)
            jj = (j*1.0/COLUMNS - 0.5)
            means_matrix[i,j] = np.sign(ii)*np.sign(jj)*np.sin((ii**2 + jj**2)/0.25)*1.0 #/(2*np.pi)
    cplot(means_matrix)
```

```python
SHUFFLE = True
np.random.seed(1310291220)
if SHUFFLE:
    row_shuffle_order = np.arange(0,ROWS,dtype=np.int)
    np.random.shuffle(row_shuffle_order)
    col_shuffle_order = np.arange(0,COLUMNS,dtype=np.int)
    np.random.shuffle(col_shuffle_order)
    shuffled_means_matrix = means_matrix[row_shuffle_order,:][:,col_shuffle_order]
else:
    shuffled_means_matrix = means_matrix

fig = plt.figure(figsize=(16,10))
fig.add_subplot(311)
cplot(means_matrix)
plt.colorbar()
plt.title("Probability Field")
fig.add_subplot(312)
cplot(shuffled_means_matrix)
plt.colorbar()
plt.title("Shuffled Probability Field")
pf = artificial_data.ProbabilityField(shuffled_means_matrix/2.0+0.5)
data = pf.realize()
fig.add_subplot(313)
bwplot(data*2-1.0)
plt.colorbar()
plt.title("Data Realization")
plt.show()
```
