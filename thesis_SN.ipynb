```python
%run py_load_data.py sn
binary_data = np.ones(data.shape)*(data > 0.0)
def remove_data(indices,direction):
    global data, binary_data
    if direction == "words":
        global words
        keep_indices = [x for x in xrange(data.shape[0]) if x not in indices]
        data = data[keep_indices,:]
        words = words[keep_indices]
        print "removed {} words".format(len(indices))
    elif direction == "docs":
        keep_indices = [x for x in xrange(data.shape[1]) if x not in indices]
        global doc_titles, doc_class
        data = data[:,keep_indices]
        doc_titles = doc_titles[keep_indices]
        doc_class = doc_class[keep_indices]
        print "removed {} docs".format(len(indices))
    binary_data = np.ones(data.shape)*(data > 0.0)
```

```python
#define some parameters that we will reuse
kwargs = {}
kwargs["threshold"] = 0.0
kwargs["row_alpha"] = 0.0
kwargs["col_alpha"] = 0.0
kwargs["row_beta"] = 1.0
kwargs["col_beta"] = 1.0
kwargs["tree_constant"] = 1.0
kwargs["n_iters"] = 3
params = questionnaire.PyQuestParams(questionnaire.INIT_AFF_COS_SIM,questionnaire.TREE_TYPE_FLEXIBLE,
                                 questionnaire.DUAL_EMD,questionnaire.DUAL_EMD,**kwargs)
params = questionnaire.PyQuestParams(questionnaire.INIT_AFF_GAUSSIAN,questionnaire.TREE_TYPE_FLEXIBLE,
                                 questionnaire.DUAL_EMD,questionnaire.DUAL_EMD,**kwargs)
```

```python
qrun = questionnaire.pyquest(binary_data,params)
row_tree = qrun.row_trees[-1]
col_tree = qrun.col_trees[-1]
row_aff = dual_affinity.emd_dual_aff(dual_affinity.calc_emd(binary_data.T,col_tree,params.row_alpha,params.row_beta))
col_aff = dual_affinity.emd_dual_aff(dual_affinity.calc_emd(binary_data,row_tree,params.col_alpha,params.col_beta))
row_vecs,row_vals = markov.markov_eigs(row_aff,12)
col_vecs,col_vals = markov.markov_eigs(col_aff,12)
```

```python
COLORS = "mgkybrcw"
#COLORS = "rbrrcbcb"
nodecolors = [COLORS[x-1] for x in doc_class]
for x,y in zip(COLORS,score_titles):
    print x,y
plot_utils.plot_embedding(row_vecs,row_vals)
plot_utils.plot_embedding(col_vecs,col_vals,nodecolors=nodecolors)
plt.show()
```
