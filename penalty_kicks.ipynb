```python
import numpy as np
import cluster_diffusion as cdiff
import qmain
import random
outcomes_matrix = np.zeros([2,2,2,2])
outcomes_matrix[:,:,0,0] = np.array([[.55,.9],[.9,.7]])
outcomes_matrix[:,:,0,1] = np.array([[.45,.9],[.9,.6]])
outcomes_matrix[:,:,1,0] = np.array([[.65,.95],[.95,.8]])
outcomes_matrix[:,:,1,1] = np.array([[.55,.95],[.95,.7]])
```

```python
#okay, four of each type in each spot: good 30/70, good 70/30, bad 30/70, bad 70/30
x_types = np.zeros([16,])
x_types[0:8] = 1
x_types[9:16] = 0

x_strats = np.zeros([16,2])
x_strats[0:4,:] = np.array([0.3,0.7])
x_strats[8:12,:] = np.array([0.3,0.7])
x_strats[4:8,:] = np.array([0.7,0.3])
x_strats[12:16,:] = np.array([0.7,0.3])
```

```python
y_strats = np.zeros([16,2])
for i in xrange(16):
    if i % 2 == 0:
        y_strats[i,:] = np.array([0.3,0.7])
    else:
        y_strats[i,:] = np.array([0.7,0.3])

y_types = np.array([0.,0.,1.,1.]*4)
```

```python
matrix = np.zeros([16,16])
for i in xrange(16):
    for j in xrange(16):
        matrix[i,j] = (x_strats[i,:].dot(outcomes_matrix[:,:,x_types[i],y_types[j]])).dot(y_strats[j,:])
```

```python
col_data = matrix
col_aff = qmain.local_geometry_norm_ip(col_data)
p,eigvals,eigvecs = cdiff.diffusion_embed(col_aff,normalized=True)
n_eigs = np.sum(np.abs(eigvals[~np.isnan(eigvals)]) > 1e-14)
tree_cols = cdiff.cluster(eigvecs[:,1:n_eigs],eigvals[1:n_eigs])
tree_cols.disp_tree()

row_data = qmain.extend_coords_means(matrix.T,tree_cols,True)
row_aff = qmain.local_geometry_norm_ip(row_data)
p,eigvals,eigvecs = cdiff.diffusion_embed(row_aff,normalized=True)
n_eigs = np.sum(np.abs(eigvals[~np.isnan(eigvals)]) > 1e-14)
tree_rows = cdiff.cluster(eigvecs[:,1:n_eigs],eigvals[1:n_eigs])
tree_rows.disp_tree()
```

```python
data = np.zeros([16,16])
data2 = np.zeros([16,16])

for n in xrange(10000):
    i = random.choice(range(16))
    j = random.choice(range(16))
    r = random.random()
    if r < matrix[i,j]:
        data[i,j] += 1
    data2[i,j] += 1
```

```python
data3 = data/data2
```

```python
col_data = data3
col_aff = qmain.local_geometry_norm_ip(col_data)
p,eigvals,eigvecs = cdiff.diffusion_embed(col_aff,normalized=True)
n_eigs = np.sum(np.abs(eigvals[~np.isnan(eigvals)]) > 1e-14)
tree_cols = cdiff.cluster(eigvecs[:,1:n_eigs],eigvals[1:n_eigs])
tree_cols.disp_tree()

row_data = qmain.extend_coords_means(data3.T,tree_cols,True)
row_aff = qmain.local_geometry_norm_ip(row_data)
p,eigvals,eigvecs = cdiff.diffusion_embed(row_aff,normalized=True)
n_eigs = np.sum(np.abs(eigvals[~np.isnan(eigvals)]) > 1e-14)
tree_rows = cdiff.cluster(eigvecs[:,1:n_eigs],eigvals[1:n_eigs])
tree_rows.disp_tree()
```

```python
for i in xrange(3):
    col_data = qmain.extend_coords_means(data3,tree_rows,True,False)
    col_aff = qmain.local_geometry_gaussian(col_data)
    p,eigvals,eigvecs = cdiff.diffusion_embed(col_aff,normalized=True)
    n_eigs = np.sum(np.abs(eigvals[~np.isnan(eigvals)]) > 1e-14)
    tree_cols = cdiff.cluster(eigvecs[:,1:n_eigs],eigvals[1:n_eigs])
    tree_cols.disp_tree()
    
    row_data = qmain.extend_coords_means(data3.T,tree_cols,True,False)
    row_aff = qmain.local_geometry_gaussian(row_data)
    p,eigvals,eigvecs = cdiff.diffusion_embed(row_aff,normalized=True)
    n_eigs = np.sum(np.abs(eigvals[~np.isnan(eigvals)]) > 1e-14)
    tree_rows = cdiff.cluster(eigvecs[:,1:n_eigs],eigvals[1:n_eigs])
    tree_rows.disp_tree()
```
