```python
data = np.ones([1000,1000],np.int)
c1 = np.random.randint(0,1000,2)
c2 = np.random.randint(0,1000,2)
data[c1[0]:c2[0],c1[1]:c2[1]] = -1
data[c1[0]:c2[0],c2[1]:c1[1]] = -1
data[c2[0]:c1[0],c1[1]:c2[1]] = -1
data[c2[0]:c1[0],c2[1]:c1[1]] = -1
row_order = np.arange(1000)
col_order = np.arange(1000)
np.random.shuffle(row_order)
np.random.shuffle(col_order)
shuffled_data = data[row_order,:][:,col_order]
```

```python
cplot(data*1.01)
plt.show()
cplot(shuffled_data*1.01)
```

```python
#define some parameters that we will reuse
kwargs = {}
kwargs["threshold"] = 0.1
kwargs["row_alpha"] = 0.5
kwargs["col_alpha"] = 0.5
kwargs["row_beta"] = 1.0
kwargs["col_beta"] = 1.0
kwargs["tree_constant"] = 0.2
kwargs["n_iters"] = 3
params = run_quest.PyQuestParams(run_quest.INIT_AFF_COS_SIM,run_quest.TREE_TYPE_FLEXIBLE,
                                 run_quest.DUAL_EMD,run_quest.DUAL_EMD,**kwargs)
```

```python
qr = run_quest.pyquest(shuffled_data,params)
```

```python
r_mat = tree_recon.recon_2d_sure(shuffled_data,qr.row_trees[-1],qr.col_trees[-1])
```
