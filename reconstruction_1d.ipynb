```python
def recon_once(f,tree):
    #shift the tree
    #tree starts at 0, and is divided into tree.size equal intervals on [0,1]
    fy = np.zeros([tree.size,1])
    #print f[0,:]
    tau = np.random.rand()
    #tau = 0.0
    #print "tau: {}".format(tau)
    fshifted = (f[0,:] + tau) % 1.0
    #print fshifted
    indices = (tree.size*fshifted).astype('int32')
    #print indices
    
    fcts = np.zeros(np.shape(fy))
    for (i,idx) in enumerate(indices):
        fcts[idx,0] += 1.0
        n = fcts[idx,0]
        #print (n-1/n)*fy[idx,0], (1/n)*f[1,i]
        fy[idx,0] = ((n-1)/n)*fy[idx,0]+(1/n)*f[1,i]
    #print f[1,:], fy

    coeffs,iters = lb.l1_bregman(cl[indices,:],fy[indices,:],1,threshold=1e-6,verbose=False)
    shiftedy = cl.dot(coeffs).ravel()
    shiftedx = np.linspace(0,1,tree.size+1)
    #print shiftedy.ravel()
    
    nx = (shiftedx - tau) % 1.0
    #print nx
    sorted_ind = nx[0:tree.size].argsort()
    #print sorted_ind
    
    y = shiftedy[sorted_ind]
    #print y
    y = np.hstack([y[-1],y[-1],y])
    #print y
    #x = shiftedx
    x = nx.copy()[0:tree.size]
    x.sort()
    x = np.hstack([[0.0],x,[1.0]])
    return x,y,tau,coeffs
    
```

```python
def plot_w_folders(x,y,f,tau):
    plt.step(x,y,color='r')
    plt.scatter(f[0,:],f[1,:],marker='o')
    leveldata = [('k','solid'),('m','solid'),('c','solid'),('g','solid'),('y','dashed'),('b','dotted')]
    plt.vlines([-tau % 1.0],0,1,linestyle=leveldata[0][1],color=leveldata[0][0],label="0")
    
    for j in xrange(1,tree.tree_depth):
        linelocs = [(-tau+(2.0*i+1.0)/(2**(j))) % 1 for i in xrange(2**(j-1))]
        #print linelocs
        plt.vlines(linelocs,0,1,linestyle=leveldata[j][1],color=leveldata[j][0],label="1/{}".format(2**j))
    leg = plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.08),
              ncol=3, fancybox=True)
    
```

```python
def sample_trees(f,tree,iters,xres=0.01):
    n = int(1.0/xres)
    new_x = np.linspace(0,1,n+1)
    new_y = np.zeros([iters,n+1])
    xhist = np.zeros([iters,tree.size+2])
    yhist = np.zeros([iters,tree.size+2])
    taulist = np.zeros(iters)
    for i in xrange(iters):
        x,y,tau,coeffs = recon_once(f,tree)
        xhist[i,:] = x
        yhist[i,:] = y
        taulist[i] = tau
        loc = 0
        for (idx,pt) in enumerate(new_x):
            while x[loc] < pt:
                loc+=1 
            new_y[i,idx] += y[loc]
    return new_x,np.mean(new_y,axis=0),xhist,yhist,new_y,taulist
   
```

```python
def sample_rtrees(f,tree,iters,xres=0.01):
    n = int(1.0/xres)
    new_x = np.linspace(0,1,n+1)
    new_y = np.zeros([iters,n+1])
    xhist = np.zeros([iters,tree.size+2])
    yhist = np.zeros([iters,tree.size+2])
    for i in xrange(iters):
        x,y,coeffs = recon_once_rshift(f,tree)
        xhist[i,:] = x
        yhist[i,:] = y
        loc = 0
        for (idx,pt) in enumerate(new_x):
            while x[loc] < pt:
                loc+=1 
            new_y[i,idx] += y[loc]
    return new_x,np.mean(new_y,axis=0),xhist,yhist,new_y
```
