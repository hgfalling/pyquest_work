Looking at sample, it seems that we can just use the second eigenvector to order the dudes for p/q. So let's do that:

```python
TREE_ALPHA = 1.0
nr = np.zeros(data.shape)
tree_constants = [0.05,0.1,0.15,0.2,0.25,0.4,0.5,0.6,0.8,1.0]
for TREE_CONSTANT in tree_constants:
    row_tree,col_tree,row_vecs,col_vecs,row_vals,col_vals = quest.pyquest_newtree(data,tree_constant=TREE_CONSTANT,row_alpha=TREE_ALPHA,col_alpha=TREE_ALPHA,beta=1.0,n_iters=3)
    new_recon = np.zeros(data.shape)
    new_coefs = tree_util.bitree_transform(data,row_tree,col_tree)
    new_recon = tree_recon.recon_2d_sure(data,row_tree,col_tree)
    nr += new_recon
nr /= len(tree_constants)
```

```python
cplot(nr)
```

```python
row_sort = np.argsort(row_vecs[:,1])
col_sort = np.argsort(col_vecs[:,1])
sorted_nr = nr[row_sort,:][:,col_sort]
cplot(sorted_nr)
```

```python
data2 = data[row_sort,:][:,col_sort]
cplot(data2*1.01)
```

```python
row_totals = np.sum(data,axis=1)
col_totals = np.sum(data,axis=0)
row_sort_totals = (-row_totals).argsort()
col_sort_totals = (-col_totals).argsort()
data3 = data[row_sort_totals,:][:,col_sort_totals]
cplot(data3*1.01)
```
