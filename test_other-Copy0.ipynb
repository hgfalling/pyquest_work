```python
def error_rate(data,recon_data):
    rd = recon_data.copy()
    rd[rd > 1.0] = 1.0
    rd[rd < -1.0] = -1.0
    return "L1: {0:12.5f} -- L2: {1:12.5f}".format(np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)) , np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)

def view_tree(tree,vecs,vals,descs=None,tree_desc=""):
    viewer_files.write_tree_viewer("temp.pickle",tree,vecs,vals,descs,tree_desc)
    %run tree_viewer.py temp.pickle
```

```python
REPEATS = 32
ROWS = 5
COLUMNS = 7
#2x2 probability field
row_blocks = [REPEATS]*ROWS
col_blocks = [REPEATS]*COLUMNS
means_matrix = np.zeros([np.sum(row_blocks),np.sum(col_blocks)])
rb = np.cumsum(np.array([0]+row_blocks))
cb = np.cumsum(np.array([0]+col_blocks))
for i in xrange(len(rb)-1):
    for j in xrange(len(cb)-1):
        means_matrix[rb[i]:rb[i+1],cb[j]:cb[j+1]] = np.random.rand()
cplot(means_matrix*2.0-1.0)
plt.show()
pf = artificial_data.ProbabilityField(means_matrix)
data = pf.realize()
bwplot(data*2-1.0)
plt.show()
```

```python
BIN_ALPHA = 1.0
TREE_CONSTANT = 0.4
TREE_BAL_CONSTANT = 2.0
SPIN_ITERS = 100
#binary trees
row_trees,col_trees = quest.pyquest_spin_bintrees(data,row_alpha=BIN_ALPHA,col_alpha=BIN_ALPHA,beta=1.0,bal_constant=TREE_BAL_CONSTANT
                                                  ,n_iters=3,n_spin=SPIN_ITERS)

row_tree,col_tree,row_vecs,row_vals,col_vecs,col_vals = quest.pyquest_newtree(data,tree_constant=TREE_CONSTANT,row_alpha=BIN_ALPHA,col_alpha=BIN_ALPHA,beta=1.0,n_iters=3)

new_recon = np.zeros(data.shape)
new_coefs = tree_util.bitree_transform(data,row_tree,col_tree)
new_recon = tree_recon.recon_2d_sure(data,row_tree,col_tree)

ind_errors = np.zeros([SPIN_ITERS,SPIN_ITERS,2])

rbin = np.zeros(data.shape)
bin_sure = np.zeros(data.shape)

for (ridx,row_tree) in enumerate(row_trees):
    for (cidx,col_tree) in enumerate(col_trees):
        bincoefs = tree_util.bitree_transform(data,row_tree,col_tree)
        recon = tree_util.inverse_bitree_transform_level(bincoefs,row_tree,col_tree,1+np.ceil(np.log2(ROWS)),1+np.ceil(np.log2(COLUMNS)))
        ind_errors[ridx,cidx,0] = error_rate(pf.means,(recon+1.0)/2.0)[2]
        rbin += recon
        recon = tree_recon.recon_2d_sure(data,row_tree,col_tree)
        ind_errors[ridx,cidx,1] = error_rate(pf.means,(recon+1.0)/2.0)[2]
        bin_sure += recon

rbin /= len(row_trees)*len(col_trees)
bin_sure /= len(row_trees)*len(col_trees)
```

```python
rt = row_trees[np.random.randint(0,SPIN_ITERS)]
ct = col_trees[np.random.randint(0,SPIN_ITERS)]
bincoefs = tree_util.bitree_transform(data,rt,ct)
sample_recon = tree_util.inverse_bitree_transform_level(bincoefs,rt,ct,1+np.ceil(np.log2(ROWS)),1+np.ceil(np.log2(COLUMNS)))

fig = plt.figure(figsize=(8,16))
fig.add_subplot(511)
cplot(means_matrix*2-1.0)
plt.title("True Means")
fig.add_subplot(512)
cplot(rbin)
plt.title("Binary Reconstruction (10x10 spin)")
fig.add_subplot(513)
cplot(bin_sure)
plt.title("Binary Reconstruction (SURE)")
fig.add_subplot(514)
cplot(new_recon)
plt.title("Single Reconstruction - New Tree")
fig.add_subplot(515)
cplot(sample_recon)
plt.title("Sample Reconstruction - Binary Tree")
plt.show()
```
