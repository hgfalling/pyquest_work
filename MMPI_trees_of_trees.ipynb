```python
from imports import *
import run_quest
import pyquest_M
```

```python
pm = pyquest_M.PyQuestDataModel()
pm.load_data("MMPI2_Depolarized.npz")
```

```python
#define some parameters that we will reuse
kwargs = {}
kwargs["threshold"] = 0.1
kwargs["row_alpha"] = 0.5
kwargs["col_alpha"] = 0.5
kwargs["row_beta"] = 1.0
kwargs["col_beta"] = 1.0
kwargs["tree_constant"] = 0.2
kwargs["n_iters"] = 5
params = run_quest.PyQuestParams(run_quest.INIT_AFF_COS_SIM,run_quest.TREE_TYPE_FLEXIBLE,
                                 run_quest.DUAL_EMD,run_quest.DUAL_EMD,**kwargs)
```

```python
qr = run_quest.pyquest(pm.data,params)
```

```python
rmat = tree_recon.recon_2d_sure(pm.data,qr.row_trees[-1],qr.col_trees[-1])
```

```python
cplot(barcode.organize_folders(qr.row_trees[-1],qr.col_trees[-1],rmat))
```

```python
rt = qr.row_trees[-1]
ct = qr.col_trees[-1]
```

```python
rmat_avgs = tree_util.bitree_averages(np.abs(rmat),rt,ct)
```

```python
total_size = rt.size * ct.size
```

```python
rc = [x.idx for x in rt.level_nodes(2)]
cc = [x.idx for x in ct.level_nodes(2)]
submodels = {}
for i in rc:
    for j in cc:
        if rmat_avgs[i,j] < 0.5:
            submodels[(i,j)] = run_quest.pyquest(pm.data[rt[i].elements,:][:,ct[j].elements],params)
```

```python
submodels.keys()
```

```python
rmat3 = rmat.copy()
for (i,j),val in submodels.items():
    subrecon = tree_recon.recon_2d_sure(pm.data[rt[i].elements,:][:,ct[j].elements],val.row_trees[-1],val.col_trees[-1])
    for ridx,row in enumerate(rt[i].elements):
        for cidx,col in enumerate(ct[j].elements):
            rmat3[row,col] = subrecon[ridx,cidx]
cplot(barcode.organize_folders(rt,ct,rmat3))
plt.show()
cplot(barcode.organize_folders(rt,ct,rmat))
```

```python
rc = [x.idx for x in rt.level_nodes(3)]
cc = [x.idx for x in ct.level_nodes(3)]
submodels = {}
for i in rc:
    for j in cc:
        if rmat_avgs[i,j] < 0.5:
            submodels[(i,j)] = run_quest.pyquest(pm.data[rt[i].elements,:][:,ct[j].elements],params)
```

```python
rmat3 = rmat.copy()
for (i,j),val in submodels.items():
    subrecon = tree_recon.recon_2d_sure(pm.data[rt[i].elements,:][:,ct[j].elements],val.row_trees[-1],val.col_trees[-1])
    for ridx,row in enumerate(rt[i].elements):
        for cidx,col in enumerate(ct[j].elements):
            rmat3[row,col] = subrecon[ridx,cidx]
cplot(barcode.organize_folders(rt,ct,rmat3))
plt.show()
cplot(barcode.organize_folders(rt,ct,rmat))
```
