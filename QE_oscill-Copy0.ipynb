```python
def thres_recon(data):
    data[data > 1.0] = 1.0
    data[data < -1.0] = -1.0

def error_rate(data,recon_data):
    rd = recon_data.copy()
    rd[rd > 1.0] = 1.0
    rd[rd < -1.0] = -1.0
    return "L1: {0:12.5f} -- L2: {1:12.5f}".format(np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)) , np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)

def view_tree(tree,vecs,vals,descs=None,tree_desc=""):
    viewer_files.write_tree_viewer("temp.pickle",tree,vecs,vals,descs,tree_desc)
    %run tree_viewer.py temp.pickle
```

```python
ROWS = 400
COLUMNS = 200
means_matrix = np.zeros([ROWS,COLUMNS])
for i in xrange(ROWS):
    for j in xrange(COLUMNS):
        ii = i*2.0*np.pi/ROWS
        jj = j*2.0*np.pi/COLUMNS
        means_matrix[i,j] = np.sin((ii+jj+2*ii*jj)/2.0)*1.0
```

```python
SHUFFLE = True
np.random.seed(52477)

row_shuffle_order = np.arange(0,ROWS,dtype=np.int)
col_shuffle_order = np.arange(0,COLUMNS,dtype=np.int)

if SHUFFLE:
    np.random.shuffle(row_shuffle_order)
    np.random.shuffle(col_shuffle_order)
else:
    shuffled_means_matrix = means_matrix

shuffled_means_matrix = means_matrix[row_shuffle_order,:][:,col_shuffle_order]
inverse_row_order = row_shuffle_order.argsort()
inverse_col_order = col_shuffle_order.argsort()

fig = plt.figure(figsize=(18,12))
#fig = plt.figure()
fig.add_subplot(311)
cplot(means_matrix)
plt.colorbar()
plt.title("Probability Field")
fig.add_subplot(312)
cplot(shuffled_means_matrix)
plt.colorbar()
plt.title("Shuffled Probability Field")
pf = artificial_data.ProbabilityField(shuffled_means_matrix/2.0+0.5)
data = pf.realize()
fig.add_subplot(313)
bwplot(data*2-1.0)
plt.colorbar()
plt.title("Data Realization")
plt.tight_layout()
plt.show()
fig.savefig("../../questionnaire_example/data.eps",format="eps")
```

```python
TREE_ALPHA = 0.5
total_recon = np.zeros(data.shape)
total_ones = np.zeros(data.shape)
subrows = int(0.8*ROWS)
subcols = int(0.8*COLUMNS)

for i in xrange(20):
    print i
    random_rows = np.arange(0,ROWS)
    np.random.shuffle(random_rows)
    random_cols = np.arange(0,COLUMNS)
    np.random.shuffle(random_cols)
    random_data = data[random_rows[0:subrows],:][:,random_cols[0:subcols]]
    
    tree_constants = [0.05,0.1,0.15,0.2,0.25,0.4,0.5,0.6,0.8,1.0]
    nr = np.zeros(random_data.shape)
    row_trees = []
    col_trees = []
    
    for TREE_CONSTANT in tree_constants:
        row_tree,col_tree,row_vecs,col_vecs,row_vals,col_vals = quest.pyquest_newtree(random_data,tree_constant=TREE_CONSTANT,row_alpha=TREE_ALPHA,col_alpha=TREE_ALPHA,beta=1.0,n_iters=3)
        row_trees.append(row_tree)
        col_trees.append(col_tree)
    
    for row_tree in row_trees:
        for col_tree in col_trees:
            new_recon = np.zeros(random_data.shape)
            new_coefs = tree_util.bitree_transform(random_data,row_tree,col_tree)
            new_recon = tree_recon.recon_2d_sure(random_data,row_tree,col_tree)
            nr += new_recon
    nr /= len(row_trees)*len(col_trees)
    nnr = np.zeros([ROWS,COLUMNS])
    nnr[0:subrows,0:subcols] += nr
    ones_nr = np.zeros([ROWS,COLUMNS])
    ones_nr[0:subrows,0:subcols] = 1.0
    nrr = nnr[random_rows.argsort(),:][:,random_cols.argsort()]
    total_ones += ones_nr[random_rows.argsort(),:][:,random_cols.argsort()]
    total_recon += nrr 
```
