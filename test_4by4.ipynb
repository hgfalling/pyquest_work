```python
def error_rate(data,recon_data):
    rd = recon_data.copy()
    rd[rd > 1.0] = 1.0
    rd[rd < -1.0] = -1.0
    return "L1: {0:12.5f} -- L2: {1:12.5f}".format(np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)) , np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)

def view_tree(tree,vecs,vals,descs=None,tree_desc=""):
    viewer_files.write_tree_viewer("temp.pickle",tree,vecs,vals,descs,tree_desc)
    %run tree_viewer.py temp.pickle
```

```python
REPEATS = 16
#2x2 probability field
row_blocks = [REPEATS]*4
col_blocks = [REPEATS]*4
means_matrix = np.zeros([np.sum(row_blocks),np.sum(col_blocks)])
rb = np.cumsum(np.array([0]+row_blocks))
cb = np.cumsum(np.array([0]+col_blocks))
for i in xrange(len(rb)-1):
    for j in xrange(len(cb)-1):
        means_matrix[rb[i]:rb[i+1],cb[j]:cb[j+1]] = np.random.rand()
bwplot(means_matrix)
plt.show()
pf = artificial_data.ProbabilityField(means_matrix)
data = pf.realize()
bwplot(data*2-1.0)
plt.show()
```

```python
BIN_ALPHA = 0.5
NEW_ALPHA = 0.5
TREE_CONSTANT = 0.5
#binary trees
bin_row_tree,bin_col_tree,bin_row_vecs,bin_col_vecs,bin_row_vals,bin_col_vals = quest.pyquest_bintree(data,row_alpha=BIN_ALPHA,col_alpha=BIN_ALPHA,beta=1.0,noise=0.0,n_iters=3)
#new trees
row_tree,col_tree,row_vecs,col_vecs,row_vals,col_vals = quest.pyquest_newtree(data,tree_constant=TREE_CONSTANT,row_alpha=NEW_ALPHA,col_alpha=NEW_ALPHA,beta=1.0,n_iters=3)
#view_tree(row_tree,row_vecs,row_vals)

bincoefs = tree_util.bitree_transform(data,bin_row_tree,bin_col_tree)
rbin = tree_util.inverse_bitree_transform_level(bincoefs,bin_row_tree,bin_col_tree,3,3)
coefs = tree_util.bitree_transform(data,row_tree,col_tree)
rdata = tree_util.inverse_bitree_transform_level(coefs,row_tree,col_tree,3,3)
bin_sure = tree_recon.recon_2d_sure(data,bin_row_tree,bin_col_tree)
new_sure = tree_recon.recon_2d_sure(data,row_tree,col_tree)

fig = plt.figure(figsize=(12,12))
fig.add_subplot(321)
bwplot(means_matrix)
plt.title("True Means")
fig.add_subplot(322)
bwplot((rbin+1.0)/2.0)
plt.title("Binary Reconstruction")
fig.add_subplot(323)
bwplot((rdata+1.0)/2.0)
plt.title("New Trees Reconstruction")
fig.add_subplot(324)
bwplot((bin_sure+1.0)/2.0)
plt.title("Binary Trees (SURE)")
fig.add_subplot(325)
bwplot((new_sure+1.0)/2.0)
plt.title("New Trees (SURE)")
plt.show()
```
