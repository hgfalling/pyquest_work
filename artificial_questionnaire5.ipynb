```python
#MEAN_CHOICES = np.linspace(0.2,0.8,num=4)
MEAN_CHOICES = np.array([0.1,0.3,0.5,0.7,0.9])

QUESTION_COUNT = 32
QUESTION_LEVELS = [2,4,4,0]
QUESTION_NOISE_BASE = 0.01
QUESTION_NOISE_ALPHA = 0.25

PEOPLE_COUNT = 64
PEOPLE_LEVELS = [4,4,4,0]
PEOPLE_NOISE_BASE = 0.01
PEOPLE_NOISE_ALPHA = 0.25

people_tree = tree.ClusterTreeNode(range(PEOPLE_COUNT))
for idx,level in enumerate(PEOPLE_LEVELS):
    for y in people_tree.dfs_level(idx+1):
        if level == 0:
            y.create_subclusters(range(y.size))
        else:
            y.create_subclusters(np.random.randint(0,level,size=y.size))
people_tree.make_index()

question_tree = tree.ClusterTreeNode(range(QUESTION_COUNT))
for idx,level in enumerate(QUESTION_LEVELS):
    for y in question_tree.dfs_level(idx+1):
        if level == 0:
            y.create_subclusters(range(y.size))
        else:
            y.create_subclusters(np.random.randint(0,level,size=y.size))
question_tree.make_index()

people_profiles = len(people_tree.dfs_level(2))
question_profiles = len(question_tree.dfs_level(2))


for node in people_tree:
    people_noise_var = PEOPLE_NOISE_BASE*(PEOPLE_NOISE_ALPHA**(node.level-2))
    if node.idx == 0:
        node.noise = np.zeros(question_profiles)
    else:
        node.noise = np.random.multivariate_normal(np.zeros(question_profiles,np.float),np.eye(question_profiles)*people_noise_var)
    if node.level == 1:
        node.mean = np.zeros(question_profiles)
    elif node.level == 2:
        node.profile = np.random.randint(0,len(MEAN_CHOICES)**question_profiles)
        node.mean = np.array(artificial_data.make_profile(node.profile,question_profiles,MEAN_CHOICES))
    else:
        node.mean = node.parent.mean + node.noise

people_means = np.vstack([node.mean for node in people_tree.leaves()]).T

for node in question_tree:
    question_noise_var = QUESTION_NOISE_BASE*(QUESTION_NOISE_ALPHA**(node.level-2))
    if node.idx == 0:
        node.noise = np.zeros(people_profiles)
    else:
        node.noise = np.random.multivariate_normal(np.zeros(people_profiles,np.float),np.eye(people_profiles)*question_noise_var)
    if node.level == 1:
        node.mean = np.zeros(people_profiles)
    else:
        node.mean = node.parent.mean + node.noise

question_noise = np.vstack([node.mean for node in question_tree.leaves()])

means = np.zeros([QUESTION_COUNT,PEOPLE_COUNT])

for question in question_tree.dfs_level(2):
    means[question.elements,:] = people_means[question.idx-1,:]
for person in people_tree.dfs_level(2):
    means[:,person.elements] += np.tile(question_noise[:,person.idx-1],(len(person.elements),1)).T

data = means.copy()
data[data > 1.0] = 1.0
data[data < 0.0] = 0.0
data = data*2 - 1.0
    
```
