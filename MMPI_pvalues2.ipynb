## Kolmogorov-Smirnov distances and analysis of MMPI Data

This notebook contains some analysis of the questionnaire process using MMPI data. I will use this type of cell to briefly explain what each code cell does. This is just importing necessary packages for the questionnaire.

```python
from imports import *
import ksmetric
import run_quest
import sklearn.decomposition as skd
```

This loads the MMPI data and the 169 questionnaire runs (saved earlier, each run has a different pair of trees generated from different parameters). Each questionnaire run is trimmed so that the predictions are capped at 1/-1. Then the entire thing is converted to probabilities, and the data is converted to 0/1 data. We take the mean of the 169 runs. (more on that later)

```python
X = (np.load("MMPI2_Depolarized.npz")["data"] + 1.0)/2.0
rhist = np.load("recon_hist.npz")['recon_hist']
rhist[rhist > 1.0] = 1.0
rhist[rhist < -1.0] = -1.0
Xhat = (np.mean(rhist,axis=0) + 1.0)/2.0
```

We calculate the KS metric for the original data and the mean reconstruction:

```python
base_ksm = ksmetric.ks_metric(X,Xhat)
print base_ksm
```

Below is a graph of the KS metric. The blue line (left scale) is the cumulative sum of the residuals -- the graph is ordered by the reconstruction probability from left to right. The red line (right scale) is the reconstruction probability at that point. So you can see that 0.5 probability comes to the right of the median (the overall average answer in the data is 0.414) And naturally the maximum (or minimum if it goes that way) of the blue line is the KS value. This corresponds to $X$ vs $\hat{X}$ in our discussion.

```python
sorder = Xhat.flatten().argsort()
cs = np.cumsum((Xhat-X).flatten()[sorder])
ax = plt.gca()
ax2 = ax.twinx()
ax.plot(cs)
ax2.plot(Xhat.flatten()[sorder],'r')
plt.show()
```

This is the KS graph of a different realization $Y$ of the probabilities in $\hat{X}$. This is NOT the right thing to do. But it gives a flavor of what the graph looks like when the true model is the source.

```python
sorder = Xhat.flatten().argsort()
Y = ksmetric.realize(Xhat)
cs = np.abs(np.cumsum(Xhat.flatten()[sorder]-Y.flatten()[sorder]))
plt.plot(cs)
plt.show()
```

Defines some set of parameters we will use to run a questionnaire on test data. These are just my favorite settings: flexible trees with 0.2 constant, $\alpha = 0.5, \beta=1.0$, and 3 iterations.

```python
kwargs = {}
kwargs["n_iters"] = 3
kwargs["threshold"] = 0.1
kwargs["row_alpha"] = 0.5
kwargs["col_alpha"] = 0.5
kwargs["row_beta"] = 1.0
kwargs["col_beta"] = 1.0
kwargs["tree_constant"] = 0.2
params = run_quest.PyQuestParams(run_quest.INIT_AFF_COS_SIM,run_quest.TREE_TYPE_FLEXIBLE,
                                 run_quest.DUAL_EMD,run_quest.DUAL_EMD,**kwargs)
```

We run a single questionnaire on the data here, and then reconstruct using SURE. We'll call that $\tilde{X}$.

```python
qr = run_quest.pyquest(X*2.0-1.0,params)
Xtilde = tree_recon.recon_2d_sure(X*2.0-1.0,qr.row_trees[-1],qr.col_trees[-1])
Xtilde[Xtilde > 1.0] = 1.0
Xtilde[Xtilde < -1.0] = -1.0
```

Okay, now here's the same KS graph for just one questionnaire run from scratch on the data, and the KS-statistic is printed above the graph. Note that the KS value is SMALLER than the one for $\hat{X}$.

```python
sorder = Xtilde.flatten().argsort()
nr = (Xtilde + 1.0)/2.0
print ksmetric.ks_metric(X,nr)
plt.plot(np.cumsum((nr-X).flatten()[sorder]))
plt.show()
```

Let's investigate this. We calculate the KS metric for the 169 different reconstructions. Amazingly, they all outperform the average of them. I want to ask Mark about this.

```python
ksms = np.zeros([13,13])
for i in xrange(13):
    for j in xrange(13):
        ksms[i,j] = ksmetric.ks_metric(X,(rhist[13*i+j,:,:]+1.0)/2.0)

print "Max: {} Min: {}".format(ksms.max(),ksms.min())
```

Here are six KS graphs for random reconstructions out of the 169. They all have the same basic shape, so that seems to be consistent.

```python
for i in np.random.randint(0,169,6):
    Xhat1 = rhist[i,:,:]
    print ksms[i/13,i%13]
    sorder = Xhat1.flatten().argsort()
    nr = (Xhat1 + 1.0)/2.0
    plt.plot(np.cumsum((nr-X).flatten()[sorder]))
    plt.show()
```

Now we will do some testing. We do the following: Realize $Y$ from $\hat{X}$. Then run the questionnaire on $Y$, and obtain $\hat{Y}$, the reconstruction of $Y$. Then calculate and plot the KS metric for $Y$ and $\hat{Y}$

```python
for i in range(6):
    new_data = ksmetric.realize(Xhat)*2.0-1.0
    qr = run_quest.pyquest(new_data,params)
    new_recon = tree_recon.recon_2d_sure(new_data,qr.row_trees[-1],qr.col_trees[-1])
    new_recon[new_recon > 1.0] = 1.0
    new_recon[new_recon < -1.0] = -1.0
    nr = (new_recon+1.0)/2.0
    nd = (new_data+1.0)/2.0
    print ksmetric.ks_metric(nd,nr)
    sorder = nr.flatten().argsort()
    plt.plot(np.cumsum((nr-nd).flatten()[sorder]))
    plt.show()
```

As a comparison, we use PCA to reconstruct the matrix. We pick a number of components and truncate the singular values above that to reconstruct.

```python
for components in [1,5,15,50,86]:
    pca = skd.PCA(n_components=components)
    pca.fit(X.T)
    pca_recon = pca.inverse_transform(pca.transform(X.T))
    ks = ksmetric.ks_metric(X,pca_recon.T)
    print "{}: {}".format(pca.n_components,ks)
    pca_recon[pca_recon > 1.0] = 1.0
    pca_recon[pca_recon < -1.0] = -1.0
    nr = (pca_recon.T+1.0)/2.0
    nd = (X+1.0)/2.0
    sorder = nr.flatten().argsort()
    plt.plot(np.cumsum((nr-nd).flatten()[sorder]))
    plt.show()
```
