```python
#ok so suppose we want to build a tree.
#well, suppose we use the labels from n=5 to do each break.
#we write a function for each break:

def break_indices(data, node, row_tree):
    #so our function will first calculate the eigenvector for the indices.
    col_indices = sorted(node.elements)
    node_data = data[:,col_indices].astype(np.float64)
    col_emd = dual_affinity.calc_emd(node_data,row_tree,0.5,1.0)
    col_aff = dual_affinity.emd_dual_aff(col_emd)
    vecs,_ = markov.markov_eigs(col_aff,2)
    eig = vecs[:,1]
    alphas,active,coefs = sklm.lars_path(node_data.T.astype(np.float64),eig)
    lm = sklm.LinearRegression()
    regressors = active[0:5]
    lm.fit(node_data.T[:,regressors],eig)
    pred_eigs = lm.predict(node_data.T[:,regressors])
    order = eig.argsort()
    mq = np.abs(eig) > 0.1
    accuracy = (np.sum((mq*pred_eigs*eig) > 0.0))/(1.0*np.sum(mq))
    labels = pred_eigs > 0.0
    partition = labels*np.ones(labels.shape[0])
    node.create_subclusters(partition)    
    return eig,regressors,lm,accuracy

def process_node(data,row_tree,node_list):
    node = node_list.pop(0)
    if node.size < 10:
        node.create_subclusters(range(node.size))
    else:
        eig,active,lm,accuracy = break_indices(data,node,row_tree)
        node.accuracy = accuracy
        node.lm = lm
        node.active = active
        node_list.extend(node.children)

def mtree(data,row_tree):
    root = tree.ClusterTreeNode(range(data.shape[1]))
    node_list = []
    node_list.append(root)
    while node_list:
        process_node(data,row_tree,node_list)
    
    root.make_index()
    return root
```

```python
def c_break_indices(data, node, row_tree, row_indices):
    #indices must be sorted!
    #so our function will first calculate the eigenvector for the indices.
    col_indices = node.elements
    node_data = data[:,col_indices].astype(np.float64)
    col_emd = dual_affinity.calc_emd(node_data,row_tree,0.5,1.0)
    col_aff = dual_affinity.emd_dual_aff(col_emd)
    vecs,_ = markov.markov_eigs(col_aff,2)
    eig = vecs[:,1]
    alphas,active,coefs = sklm.lars_path(node_data[row_indices,:].T.astype(np.float64),eig,max_iter=15)
    lm = sklm.LinearRegression()
    regressors = active[0:5]
    lm.fit(node_data.T[:,regressors],eig)
    pred_eigs = lm.predict(node_data.T[:,regressors])
    order = eig.argsort()
    mq = np.abs(eig) > 0.1
    accuracy = (np.sum((mq*pred_eigs*eig) > 0.0))/(1.0*np.sum(mq))
    labels = pred_eigs > 0.0
    partition = labels*np.ones(labels.shape[0])
    node.create_subclusters(partition)    
    return eig,regressors,lm,accuracy

def c_process_node(data,row_tree,node_list,row_indices):
    node = node_list.pop(0)
    if node.size < 10:
        node.create_subclusters(range(node.size))
    else:
        eig,active,lm,accuracy = c_break_indices(data,node,row_tree,row_indices)
        node.accuracy = accuracy
        node.lm = lm
        node.active = active
        node_list.extend(node.children)

def c_mtree(data,row_tree,row_indices):
    root = tree.ClusterTreeNode(range(data.shape[1]))
    node_list = []
    node_list.append(root)
    while node_list:
        c_process_node(data,row_tree,node_list,row_indices)
    
    root.make_index()
    return root
```
