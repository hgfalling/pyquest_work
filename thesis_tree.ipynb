We generate a toy affinity for demonstration purposes by taking $A(i,j) \sim \exp{\left(-\frac{\left|{x_i-x_j}\right|^2}{25}\right)}$ and symmetrizing it by taking $\frac{1}{2}\left(A+A^T\right)$.

```python
r = np.zeros([64,64])
for i in xrange(64):
    for j in xrange(64):
        d_ij = (i-j)
        r[i,j] = np.exp(-(d_ij**2.0/25.0))
data = (r+r.T)/2.0
```

```python
vecs,vals = markov.markov_eigs(data,8)
```

```python
plot_utils.plot_embedding(vecs,vals,1.0)
plt.show()
```

```python
row_order = np.random.rand(64).argsort()
permuted_r = r[row_order,:][:,row_order]
bwplot(r)
plt.show()
bwplot(permuted_r)
plt.show()
```

```python
bt1 = bin_tree_build.bin_tree_build(r)
leafcolors1 = np.arange(-1.0,1.0,2.0/64.0)
colors1 = np.concatenate([np.array([-2.0]*63),leafcolors1])
plot_utils.plot_tree(bt1,colors1)
plt.show()

bt2 = bin_tree_build.bin_tree_build(permuted_r)
leafcolors2 = np.arange(-1.0,1.0,2.0/64.0)
colors2 = np.concatenate([np.array([-2.0]*63),leafcolors2[row_order]])
plot_utils.plot_tree(bt2,colors2)
plt.show()
```
