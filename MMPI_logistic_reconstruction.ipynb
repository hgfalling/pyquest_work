```python
pm = pym.PyQuestDataModel()
pm.load('base.session')
```

```python
qrun = pm.quest_runs[-1]
row_tree = qrun.row_trees[-1]
col_tree = qrun.col_trees[-1]
```

```python
plot_utils.plot_tree(row_tree)
plt.show()
plot_utils.plot_tree(col_tree)
plt.show()
```

```python
_,parents = haar.bihaar_transform(np.zeros([567,2428]),row_tree,col_tree,True)
```

```python
row_haar = haar.compute_haar(row_tree,False,'L1')
col_haar = haar.compute_haar(col_tree,False,'L1')
```

```python
def msp(psize):
    pairs = zip(*np.where(parents > psize))
    print len(pairs)
    row_data = []
    col_data = []
    data = []
    for idx,(i,j) in enumerate(pairs):
        if idx == 0:
            row_data.extend(range(567*2428))
            col_data.extend([0]*(567*2428))
            data.extend([1]*(567*2428))
        else:
            hpluses = np.where(row_haar[:,i]>0)[0]
            hminuses = np.where(row_haar[:,i]<0)[0]
            cpluses = np.where(col_haar[:,j]>0)[0]
            cminuses = np.where(col_haar[:,j]<0)[0]
            hplus = row_haar[hpluses[0],i]
            if i>0:
                hminus = row_haar[hminuses[0],i]
            cplus = col_haar[cpluses[0],j]
            if j>0:
                cminus = col_haar[cminuses[0],j]
            for (x,y) in itertools.product(hpluses,cpluses):
                row_data.append(x*2428+y)
                col_data.append(idx)
                data.append(hplus*cplus)
            for (x,y) in itertools.product(hpluses,cminuses):
                row_data.append(x*2428+y)
                col_data.append(idx)
                data.append(hplus*cminus)
            for (x,y) in itertools.product(hminuses,cpluses):
                row_data.append(x*2428+y)
                col_data.append(idx)
                data.append(hminus*cplus)
            for (x,y) in itertools.product(hminuses,cminuses):
                row_data.append(x*2428+y)
                col_data.append(idx)
                data.append(hminus*cminus)
    return spss.coo_matrix((data,(row_data,col_data)),shape=(567*2428,len(pairs)))
```

```python
%prun ind_fcn = msp(0.001)
ind_csr = ind_fcn.tocsr()
```

```python
%run py_load_data.py de
```

```python
lr = sklm.LogisticRegression()
lr.fit(ind_csr,data.flatten())
```

```python
pf = lr.predict_proba(ind_csr)
pf2 = -1.0*pf[:,0] + 1.0*pf[:,2]
pfr = np.reshape(pf2,[567,2428])
```

```python
recon = tree_recon.recon_2d_haar_folder_size(data,row_tree,col_tree,0.001)
recon = tree_recon.threshold_recon(recon,-1.0,1.0)
```
