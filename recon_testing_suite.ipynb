```python
def error_rate(data,recon_data):
    rd = recon_data.copy()
    rd[rd > 1.0] = 1.0
    rd[rd < -1.0] = -1.0
    return "L1: {0:12.5f} -- L2: {1:12.5f}".format(np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)) , np.sum(np.abs(rd-data)),np.linalg.norm(rd-data)
```

```python
#thresholding based on folder size (2d)
n_runs = 50
bin_trecon = {}
trecon = {}
terror = np.zeros([3,n_runs])
for idx,i in enumerate(np.logspace(-4.5,-1.5,num=n_runs)):
    bin_trecon[idx] = tree_recon.recon_2d_folder_size(data,bin_row_tree,bin_col_tree,i)
    trecon[idx] = tree_recon.recon_2d_folder_size(data,row_tree,col_tree,i)
    terror[0,idx] = i
    terror[1,idx] = error_rate(aq.data,bin_trecon[idx])[2]
    terror[2,idx] = error_rate(aq.data,trecon[idx])[2]
    #print "{0:9.7f}".format(i), terror[1,idx], terror[2,idx]
```

```python
plt.semilogx(terror[0,:],terror[1,:],'r',label='Binary eigentrees')
plt.semilogx(terror[0,:],terror[2,:],'b',label='New trees')
plt.xlabel('Folder size threshold')
plt.ylabel('L2 error')
plt.legend()
plt.show()
```

```python
#thresholding based on hypothesis testing
n_runs = 20

bin_pv = tree_recon.calc_hg_p_values_bitree(data,bin_row_tree,bin_col_tree)
pv = tree_recon.calc_hg_p_values_bitree2(data,row_tree,col_tree)
bin_coefs = tree_util.bitree_transform(data,bin_row_tree,bin_col_tree)
coefs = tree_util.bitree_transform(data,row_tree,col_tree)

bin_precon = {}
precon = {}
perror = np.zeros([3,n_runs])
for idx,alp in enumerate(np.logspace(-4,-1,num=n_runs)):
    bin_precon[idx] = tree_util.inverse_bitree_transform(bin_coefs*(bin_pv < alp),bin_row_tree,bin_col_tree)
    precon[idx] = tree_util.inverse_bitree_transform(coefs*(pv < alp),row_tree,col_tree)
    perror[0,idx] = alp
    perror[1,idx] = error_rate(aq.data,bin_precon[idx])[2]
    perror[2,idx] = error_rate(aq.data,precon[idx])[2]
    #print "{0:9.7f}".format(alp), perror[1,idx], perror[2,idx]
```

```python
plt.semilogx(perror[0,:],perror[1,:],'r',label='Binary eigentrees')
plt.semilogx(perror[0,:],perror[2,:],'b',label='New trees')
plt.xlabel('Significance level')
plt.ylabel('L2 error')
plt.legend()
plt.show()
```

```python
#virtual shrinkage by changing significant levels
bin_pprecon = np.zeros(data.shape)
pprecon = np.zeros(data.shape)
for i in bin_precon.values():
    bin_pprecon += i
for i in precon.values():
    pprecon += i
pprecon /= n_runs*1.0
bin_pprecon /= n_runs*1.0
bin_pperror = error_rate(aq.data,bin_pprecon)
pperror = error_rate(aq.data,pprecon)
print "binary: {} newtree: {}".format(bin_pperror[0],pperror[0])
```
