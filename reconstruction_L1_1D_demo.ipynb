```python
#divide up [0,1] into a binary tree of 5 levels (so 32 leaf folders)
tree = cdiff.dyadic_tree(5)
tree.calc_delta_library()

#generate random reproducible
seed = 20130528
f = recon.random_function_data(seed,[10,2,6])

F_TO_PROCESS = 9
print f[F_TO_PROCESS]


recon.scatter(f[F_TO_PROCESS],plt)
plt.title('random function f')
plt.ylim((-0.2,1.2))
plt.xlim((-0.2,1.2))
plt.show()
```

```python
tau = 0.0
x,y,coeffs = recon.reconstruct(f[F_TO_PROCESS],tree,tau,0.5,0.5)
plt.step(x,y,'r')
lines = recon.plot_folder_boundaries(tree,tau,plt)
recon.scatter(f[F_TO_PROCESS],plt)
plt.title('Example L1-reconstruction')
plt.show()
```

```python
N = 100
xhist,yhist = recon.sample_recon(f[F_TO_PROCESS],tree,N,0.25,0.75)
x,y = recon.combine(xhist,yhist,0.001)
```

```python
plt.step(x,np.mean(y,axis=0),'r')
plt.title('L1-reconstruction (mean of {} trees)'.format(N))
recon.scatter(f[F_TO_PROCESS],plt)
plt.show()

plt.step(x,np.median(y,axis=0),'g')
plt.title('L1-reconstruction (median of {} trees)'.format(N))
recon.scatter(f[F_TO_PROCESS],plt)
plt.show()

plt.step(x,recon.threshold_mean(y,0.8),'c')
plt.title('L1-reconstruction (mean of {} trees/thres = 0.8)'.format(N))
recon.scatter(f[F_TO_PROCESS],plt)
plt.show()

plt.step(x,recon.threshold_mean(y,0.9),'c')
plt.title('L1-reconstruction (mean of {} trees/thres = 0.9)'.format(N))
recon.scatter(f[F_TO_PROCESS],plt)
plt.show()
```

```python
xvals = [0.47+(0.01*i) for i in xrange(4)]
for xval in xvals:
    h,idx = recon.histogram_x(x,y,xval,plt)
    ybar = np.mean(y[:,idx])
    yhat = recon.threshold_mean(y[:,idx],0.8)[0]
    print h[0]
    plt.title(r'histogram at x = {0}, $\bar{1} = {2:.4f}, \hat{3} = {4:.4f}$'.format(xval,'{y}',ybar,'{y}',yhat))
    plt.show()
```

```python
sigmas = [0.02,0.05,0.1,0.15]
for sigma in sigmas:
    xk,yk = recon.gaussian_kernel_smooth(f[F_TO_PROCESS],sigma,0.001)
    plt.plot(xk,yk,'r')
    recon.scatter(f[F_TO_PROCESS],plt)
    plt.title('Gaussian kernel smoothing ($\epsilon = {}$)'.format(sigma))
    plt.show()
```
