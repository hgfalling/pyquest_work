```python
#sample helix for the x-direction:
revs = 10
t = np.random.rand(500)*revs*np.pi
x_pts = np.vstack([np.cos(t),np.sin(t), t/(revs*np.pi)])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
#ax.plot(xs=x_pts[0,:],ys=x_pts[1,:],zs=x_pts[2,:])
ax.scatter3D(xs=x_pts[0,:],ys=x_pts[1,:],zs=x_pts[2,:],c='r')
```

```python
#sample paraboloid for the y direction:
s = np.random.rand(500)
theta = np.random.rand(500)*2*np.pi
y_pts = np.vstack([s*np.cos(theta)+1.0,s*np.sin(theta)+1.0,1-s*s])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
#ax.plot(xs=y_pts[0,:],ys=y_pts[1,:],zs=y_pts[2,:])
ax.scatter3D(xs=y_pts[0,:],ys=y_pts[1,:],zs=y_pts[2,:])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")
ax.scatter3D(xs=x_pts[0,:],ys=x_pts[1,:],zs=x_pts[2,:],c='r')
ax.scatter3D(xs=y_pts[0,:],ys=y_pts[1,:],zs=y_pts[2,:],c='b')
```

```python
m = x_pts.shape[1]
n = y_pts.shape[1]
potential = np.zeros([m,n])
for i in xrange(n):
    for j in xrange(n):
        potential[i,j] = 1.0/(np.linalg.norm(x_pts[:,i] - y_pts[:,j]))
```

```python
cplot(potential)
```

```python
q_descs = [str(x) for x in range(potential.shape[0])]
p_score_descs = None
p_scores = None
write_data_file("spiral.npz",data=potential,q_descs=q_descs,p_score_descs=p_score_descs,p_scores=p_scores)
```
