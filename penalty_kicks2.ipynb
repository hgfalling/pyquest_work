```python
import numpy as np
import cluster_diffusion as cdiff
import qmain
import random
import matplotlib.pyplot as plt
import pk
import tree_utils
reload(pk)
outcomes_matrix = np.zeros([2,2,2,2])
outcomes_matrix[:,:,0,0] = np.array([[.55,.9],[.9,.7]])
outcomes_matrix[:,:,0,1] = np.array([[.45,.9],[.9,.6]])
outcomes_matrix[:,:,1,0] = np.array([[.65,.95],[.95,.8]])
outcomes_matrix[:,:,1,1] = np.array([[.55,.95],[.95,.7]])

#okay, four of each type in each spot: good 30/70, good 70/30, bad 30/70, bad 70/30
x_types = np.zeros([16,])
x_types[0:8] = 1
x_types[9:16] = 0

x_strats = np.zeros([16,2])
x_strats[0:4,:] = np.array([0.3,0.7])
x_strats[8:12,:] = np.array([0.3,0.7])
x_strats[4:8,:] = np.array([0.7,0.3])
x_strats[12:16,:] = np.array([0.7,0.3])

y_strats = np.zeros([16,2])
for i in xrange(16):
    if i % 2 == 0:
        y_strats[i,:] = np.array([0.3,0.7])
    else:
        y_strats[i,:] = np.array([0.7,0.3])

y_types = np.array([0.,0.,1.,1.]*4)

matrix = np.zeros([16,16])
for i in xrange(16):
    for j in xrange(16):
        matrix[i,j] = (x_strats[i,:].dot(outcomes_matrix[:,:,x_types[i],y_types[j]])).dot(y_strats[j,:])

n_samples = 1000

s_matrix = np.zeros([16,16])
for i in xrange(16):
    for j in xrange(16):
        s_matrix[i,j] = np.random.binomial(n_samples,matrix[i,j])/(1.0*n_samples)
        #stddev of this matrix is about .068 (for 75% cells)

r_matrix = np.zeros([16,16])

iters = 1

for i in xrange(iters):
    tree_rows,tree_cols,row_eigvecs,col_eigvecs = pk.pk_questionnaire(s_matrix,3)
    bt_coefs = tree_utils.bitree_transform(s_matrix,tree_rows,tree_cols)
    r_matrix_i, used_coefs = tree_utils.inverse_bitree_transform(bt_coefs,tree_rows,tree_cols,0.005)
    r_matrix += r_matrix_i


r_matrix /= iters
```
