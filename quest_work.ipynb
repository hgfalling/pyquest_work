```python
if 1 == 1:
    ROWS = 400
    COLUMNS = 200
    means_matrix = np.zeros([ROWS,COLUMNS])
    for i in xrange(ROWS):
        for j in xrange(COLUMNS):
            ii = i*2.0*np.pi/ROWS
            jj = j*2.0*np.pi/COLUMNS
            means_matrix[i,j] = np.sin((ii+jj+2*ii*jj)/2.0)*1.0
    cplot(means_matrix)
```

```python
SHUFFLE = True
if SHUFFLE:
    row_shuffle_order = np.arange(0,ROWS,dtype=np.int)
    np.random.shuffle(row_shuffle_order)
    col_shuffle_order = np.arange(0,COLUMNS,dtype=np.int)
    np.random.shuffle(col_shuffle_order)
    means_matrix = means_matrix[row_shuffle_order,:][:,col_shuffle_order]

pf = artificial_data.ProbabilityField(means_matrix/2.0+0.5)
data = pf.realize()
bwplot(data*2-1.0)
plt.show()
```

```python
init_row_aff = affinity.mutual_cosine_similarity(data.T,False,0,threshold=0.1)
init_row_vecs,init_row_vals = markov.markov_eigs(init_row_aff, 12)
init_row_vals[np.isnan(init_row_vals)] = 0.0
row_embedding = init_row_vecs.dot(np.diag(init_row_vals))
row_distances = spsp.distance.squareform(spsp.distance.pdist(row_embedding))
row_affinity = np.max(row_distances) - row_distances
```
