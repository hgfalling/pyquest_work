```python
import numpy as np
import cluster_diffusion as cdiff
reload(cdiff)
import l1_bregman as lb
reload(lb)
import sklearn.linear_model as sklm
import recon
reload(recon)
import matplotlib.pyplot as plt
```

```python
#generate random function data as usual
seed = 20130516
random_data = recon.random_function_data(seed,[10,2,6])

P_NO = 8 
f = recon.fsort(random_data[P_NO])
#print random_data[P_NO]
#print f

recon.scatter(f,plt)
plt.title('random function f')
plt.ylim((-0.05,1.05))
plt.xlim((-0.05,1.05))
plt.show()

tree = cdiff.dyadic_tree(5)
tree.calc_delta_library()
cl = tree.char_library()
```

```python
ALPHA = 0.8
cl = tree.char_library(ALPHA)
x_grid, fy, alphas, active_vars, coef_path, indices = recon.reconstruct_l2(f,tree,0.25,0.75,ALPHA,suppress_warnings=True)
f_interpolates = cl.dot(coef_path)
m,n = np.shape(f_interpolates)
l1_norms = np.sum(np.abs(coef_path),axis=0)
l2_norms = np.zeros(np.shape(l1_norms))

for i in xrange(n):
    l2_norms[i] = 1.0/12.0 * (np.linalg.norm(fy[indices,:] - f_interpolates[indices,i]))**2

fig = plt.figure(figsize=(16,12))

for i in xrange(n):
    plt.subplot(n/3+1,3,i+1)
    plt.xlim([-0.05,1.05])
    plt.ylim([-0.05,1.05])
    recon.scatter(f,plt)
    yvals = f_interpolates[:,i]
    x = np.hstack([x_grid,[1.0]])
    y = np.hstack([yvals[0],yvals])
    plt.step(x,y,'r')
    plt.title(r'{0} coeffs, $\alpha = {1:.4f} $: L1={2:.5f} L2={3:.5f}'.format(i,alphas[i],l1_norms[i],l2_norms[i]))
    #lines = recon.plot_folder_boundaries(tree,0.0,plt)
plt.show()    
```
